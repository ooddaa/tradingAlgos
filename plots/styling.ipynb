```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# shared y axis https://matplotlib.org/3.3.3/gallery/subplots_axes_and_figures/shared_axis_demo.html

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)

d = np.arange(0,100,1)
ax1.plot(d, d)
ax2.plot(d**2, d)
ax3.plot(d**3, d)

# remove ticks from the rightmost axis
plt.tick_params(left=False)
# remove yticklabels 
# plt.setp(ax2.get_yticklabels(), visible=False)
# plt.setp(ax3.get_yticklabels(), visible=False)


ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['bottom'].set_color('#DDDDDD')

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['bottom'].set_color('#DDDDDD')

# ax2.yaxis.grid(True, color='#AAAAAA')

ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_visible(False)
ax3.spines['bottom'].set_color('#DDDDDD')

# remove y axes
ax2.get_yaxis().set_visible(False)
ax3.get_yaxis().set_visible(False)

# plt.xticks(rotation=90) # modifies only ax2

plt.show()
```

```python
y = np.arange(0,100,1)
d1 = y
d2 = y**2
d3 = y**3
ax1 = plt.subplot(311)
ax1.plot(y, d1)
ax2 = plt.subplot(312, sharey=ax1)
ax2.plot(y, d2)
ax3 = plt.subplot(313, sharey=ax1)
ax3.plot(y, d3)
```

```python
plt.title('Expenditure growth over 2010-2019')
plt.ylabel('Per cent')
plt.legend()
plt.show()
```
