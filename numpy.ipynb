```python
# https://numpy.org/doc/stable/reference/generated/numpy.linspace.html
import matplotlib.pyplot as plt
N = 8
y = np.zeros(N)
x1 = np.linspace(0, 10, N, endpoint=True)
x2 = np.linspace(0, 10, N, endpoint=False)
x3 = np.linspace(0, 10, 7, endpoint=True)
x4 = np.linspace(0, 10, 6, endpoint=True)
x5 = np.linspace(0, 10, 5, endpoint=True)
plt.plot(x1, y, 'o')
# [<matplotlib.lines.Line2D object at 0x...>]
plt.plot(x2, y - 0.2, 'o')
# [<matplotlib.lines.Line2D object at 0x...>]
plt.plot(x3, np.zeros(7) + 0.2, 'o')
plt.plot(x4, np.zeros(6) + 0.4, 'o')
plt.plot(x5, np.zeros(5) + 0.6, 'o')

plt.ylim([-0.5, 1])
(-0.5, 1)
plt.show()
```

```python
# meshgrid
# https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html
nx, ny = (3, 2)
x = np.linspace(0, 1, nx)
y = np.linspace(0, 1, ny)
xv, yv = np.meshgrid(x, y)
# xv
# array([[0. , 0.5, 1. ],
#        [0. , 0.5, 1. ]])
# yv
# array([[0.,  0.,  0.],
#        [1.,  1.,  1.]])
xv, yv = np.meshgrid(x, y, sparse=True)  # make sparse output arrays
# xv
# array([[0. ,  0.5,  1. ]])
# yv
# array([[0.],
#        [1.]])
```

```python
# p93 Operations between differently sized arrays is called broadcasting
arr = np.array([1,2,3,4,5])
arr_slice = arr[1:3] # view, not a copy
arr_slice[:] = 666

# copy explicitly
arr_copy = arr[3:].copy()
arr_copy[:] = 123 # does not affect arr
```

```python
# indexing
arr = np.arange(0,9).reshape(3,3)
arr[0,:] # array([0, 1, 2])
arr[:1,:] # array([[0, 1, 2]])
arr[-1:, -1: ] # array([[8]])
arr[2, 2:] # array([8])
arr[2,2] # 8
arr[1,1] # 4
```

```python
# boolean indexing
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
data = np.random.randn(7,4)
print(data)
data[names == 'Bob', 1] # pick 0 and 3 rows from data and 1 elm
data[~(names == 'Bob') & ~(names == 'Will'), :2] # not Bob and not Will, first two columns
```

```python
# Vectorize array
strikes = np.array([50, 55, 60, 60, 60])
# I want to generate normally distributed random volumes over these strikes
# https://stackoverflow.com/questions/35215161/most-efficient-way-to-map-function-over-numpy-array
volumes = lambda t: np.random.normal(t*100, 1000, 100)
vfunc = np.vectorize(volumes)
s = np.random.normal(50*100, 1000, 100)
plt.hist(s, 100)
```

```python
# Normal Distribution
# https://numpy.org/devdocs/reference/random/generated/numpy.random.normal.html
mu, sigma = 0, 0.1 
s = np.random.normal(mu, sigma, 1000)

count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```python
# natural log - e based
# https://numpy.org/doc/stable/reference/generated/numpy.log.html
np.log(2.73) 

# additive
# https://mathbabe.org/2011/07/24/measuring-historical-volatility/
prices = [1,2,3,4,5,6]
logs = np.log(prices)
logs[2] + 0.01 - 0.01 == logs[2] # True
prices[2] * 1.01 * .99 == prices[2] # False
```

```python
# main reason we use logs - to compare highly distributed datasets, to shrink it, 
# turn LARGE y-axis into much more compact, normalized one

# log10
ds1 = np.arange(1, 100, 1)
ds2 = np.arange(1, 10000, 100)
plt.plot(np.log10(ds1))
plt.plot(np.log10(ds2), 
         color='red',   
         linewidth=1.0,  
         linestyle='--' 
        )

plt.show()
```

```python
# naturnal log
ds1 = np.arange(1, 100, 1)
ds2 = np.arange(1, 10000, 100)
plt.plot(np.log(ds1))
plt.plot(np.log(ds2), 
         color='red',   
         linewidth=1.0,  
         linestyle='--' 
        )

plt.show()
```
