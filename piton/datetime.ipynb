```python
# is string a valid iso date format?
def datetime_valid(dt_str):
    try:
        datetime.fromisoformat(dt_str)
    except:
        return False
    return True

datetime_valid('2020-01-01') # true
datetime_valid('0020-01-01') # true
```

```python
# I want to filter by time horizon
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html
start = datetime.now()
end = start + timedelta(400)
# print(start)
# print(end)
index = pd.date_range(start, end)
df = pd.DataFrame(np.arange(0, len(index),1), index=index)
df
```

```python
# now just filter first full year
one_year = datetime.now() + timedelta(364)
one_year_df = df[df.index < one_year]
one_year_df.shape
```

```python
# date index with duplicate dates
df1 = pd.concat([df, df], axis=0)
one_year = datetime.now() + timedelta(364)
one_year_df1 = df1[df1.index < one_year]
one_year_df1.shape
```
