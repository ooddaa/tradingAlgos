```python
# let's say we chose Bid Price and Volume as key columns to work with
# we want to know how many lines with either BP and/or V == zero
cleaned = data.loc[(data['Bid Price'] != 0) & (data['Volume'] !=0)]
count = len(cleaned)
print('rows dropped  :', data.shape[0] - count)
print('rows dropped %:', (1 - count/data.shape[0])*100)
cleaned
```

```python
# clean data

# find how many 0 volume datapoints there are
volume = data['Volume']
zero_vol = len(volume.loc[volume <= 0])
print('zero volume       :', zero_vol)
print('ok.      volume  %:', (1 - zero_vol/len(volume))*100)

# remove all zero volumes
start_shape = data.shape
dataB = data.loc[data['Volume'] > 0]
print('data points loss  :', start_shape[0] - dataB.shape[0])
print('data points loss %:', (1 - dataB.shape[0]/start_shape[0])*100)
```

```python
# clean data

# find how many 0 Last Price datapoints there are
last_price = data['Last Price']
zero_last_price = len(last_price.loc[last_price <= 0])
print('zero Last Price   :', zero_last_price)
print('ok   last_price  %:', (1 - zero_last_price/len(last_price))*100)

```

```python
# check out year 2020
data_2020 = data['20200101':].loc[data['Last Price'] != 0].describe()
mean, std = data_2020.loc[['mean','std'],'Last Price']
# Minimum price fluctuation
# 0.005 = C$12.50 per contract for all contract months
# Contract spec - https://m-x.ca/produits_taux_int_bax_en.php
price_inc = 0.005 # for each 0.005 increase in contract price
price_inc_cad = 12.5 # contract value increases CAD12.5
sigma_in_cad = (std/.005)*price_inc_cad # therefore each sigma represents CAD149.42
two_sigmas_in_cad = sigma_in_cad * 2
two_sigmas_in_cad # CAD298.85 - 95.4% of 2020 the price could fluctuate from -300 to +300 CAD
three_sigmas_in_cad = sigma_in_cad * 3 # CAD448.27
```
