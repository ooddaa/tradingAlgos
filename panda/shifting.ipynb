```python
df_date = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami', 'corned beef', 'Bacon', 'pastrami', 'honey ham', 'nova lox']}, 
                       index=pd.date_range("2021-02-01", "2021-02-09"))
df_date
```

```python
# leaves NaNs
df_date.shift(periods=2)
```

```python
# leaves NaNs
df_date.shift(periods=2, fill_value=0)
```

```python
# doesnt leave Nans
df_date.shift(periods=2, freq='D')
```

```python
# shift allows us to make a column that refers to previous days/time periods
df_date['2_days_ago'] = df_date['food'].shift(2)
df_date
```
