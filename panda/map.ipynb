```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami', 'corned beef', 'Bacon', 'pastrami', 'honey ham', 'nova lox'],
                  'ounces': [4, 3, 12, 6 ,7, 8, 3, 5, 6]})
df
```

```python
meat_to_animal = {
      'bacon': 'pig',
      'pulled pork': 'pig',
      'pastrami': 'cow',
      'corned beef': 'cow',
      'honey ham': 'pig',
      'nova lox': 'salmon'
}
```

```python
# coz foods are of different cases => we have NaNs obviously
df['animal'] = df['food'].map(meat_to_animal)
df
```

```python
# lowercase 
df['animal'] = df['food'].str.lower().map(meat_to_animal)
df
```

```python
# same as
df['animal2'] = df['food'].map(lambda x: meat_to_animal[x.lower()])
df
```
