```python
# make a df with each major >5kgbp items represented as % of total
new = bc / bc.sum() # ok
new
```

```python
items = [
    'Management Fee',
    'Insurance',
    'Reserve Fund',
    'Electricity',
    'General Repair & Maintenance',
    'Heating Plant (Boiler Maintenance)',
    'Lift Repair and Maintenance',
    'Gas (Common parts/Sports Centre)',
    'CCTV, door entry & sliding doors',
    'Garden Maintenance',
    'Fire equipment maintenance',
]

def get_results(df, items):
    """
    Returns a dict, consumable by my_hbar_plotter.
    Df cols turn into y-axis
    """
    results = {}
    for col in df:
        results[col] = []
        for item in items:
            results[col].append(df.loc[item,col])

    return results
# print(results)

my_hbar_plotter(get_results(new, items), items, figsize=(30,12))
plt.show()
```
