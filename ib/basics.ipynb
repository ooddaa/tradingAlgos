```python
def print_position(p):
    """
    account symbol currency position avgCost (currency position*avgCost)
    """
    account = p[0]
    symbol = p[1].symbol
    currency = p[1].currency
    position = p[-2]
    avgCost = p[-1]
    return f"{account} {symbol} {currency}{position} ({currency}{position*avgCost})"
```

```python
def positions_to_df(positions):
    """
    Wraps IB's Positions[] in a DataFrame
    """
    contract = [p[1].dict() for p in positions]
    df = pd.DataFrame([[p[0], p[-2], p[-1]] for p in positions], 
                      columns=['account', 'position', 'avgCost'])
    df = pd.concat([df, pd.DataFrame(contract)], axis=1)
    return df
```
