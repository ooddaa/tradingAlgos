```python
def ib_options_to_df(options, und_price):
    """
    Takes IB Ticker and returns a df.
    """
    # yahoo columns
    columns = ['contractSymbol', 'lastTradeDate', 'strike', 'lastPrice', 
               'bid', 'ask', 'change', 'percentChange', 'volume', 
               'openInterest','impliedVolatility', 'inTheMoney', 
               'contractSize', 'currency']
    
    def in_the_moneyness(strike, right, und_price):
        if right == 'p':
            return True if strike > und_price else False
        else:
            return True if strike < und_price else False
        
    def option_to_array(option, und_price):
        contractSymbol = option['contract'].localSymbol.replace(' ','')
        lastTradeDate = np.nan
        strike = option['contract'].strike
        right = option['contract'].right
        lastPrice = option['last']
        bid = option['bid']
        ask = option['ask']
        change = np.nan
        percentChange = np.nan
        
        def get_volume(option, right):
            if not np.isnan(option['volume']):
                return option['volume']
            return option['putVolume'] if right == 'P' else option['callVolume']
            
        
        openInterest = option['putOpenInterest'] if right == 'P' else option['callOpenInterest']
        impliedVolatility = option['impliedVolatility']
        inTheMoney = in_the_moneyness(strike, right, und_price)
        contractSize = 'REGULAR' if option['contract'].multiplier == '100' else 'IRREGULAR'
        currency = option['contract'].currency
        return [contractSymbol, 
                lastTradeDate, 
                strike, 
                lastPrice,
                bid,
                ask,
                change,
                percentChange,
                get_volume(option, right),
                openInterest,
                impliedVolatility,
                inTheMoney,
                contractSize,
                currency
               ]
    
    data = [option_to_array(option, und_price) for option in options]
    result = pd.DataFrame(data, columns=columns)
    return result
```
